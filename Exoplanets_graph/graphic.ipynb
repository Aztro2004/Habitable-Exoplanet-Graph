```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd 
```

```python
data = pd.read_csv('EXOPLANETAS.csv')
```

```python
data.columns
```

```python
radius_planet = data['pl_rade']
radius_sun = data['st_rad']  
radius_orb  = data ['pl_orbsmax']
```

```python
def Sun_planet_orb(radius_planet, radius_sun, radius_orb=None):

    if radius_orb is None:
        radius_orb = 1 + (radius_sun * 0.20)

    num_points = 1000  

    # Create the orbit path
    theta = np.linspace(0, 2*np.pi, num_points)
    x_orbit = radius_orb * np.cos(theta)
    y_orbit = radius_orb * np.sin(theta)
    z_orbit = np.zeros_like(x_orbit)

    # Create the planet
    x_planet = [x_orbit[0]]
    y_planet = [y_orbit[0]]
    z_planet = [z_orbit[0]]

    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    ax.scatter(0, 0, 0, color='yellow', label='Sun',s=radius_sun*100)

    ax.plot(x_orbit, y_orbit, z_orbit, color='white', linestyle='dashed', label='Orbit')

    ax.scatter(x_planet, y_planet, z_planet, color='blue', label='Planet',s=radius_planet*100)

    ax.set_xlim(-radius_orb-(radius_orb*.3),radius_orb+(radius_orb*.3))
    ax.set_ylim(-radius_orb-(radius_orb*.3),radius_orb+(radius_orb*.3))
    ax.set_zlim(-radius_orb-(radius_orb*.3),radius_orb+(radius_orb*.3))
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.set_title('Planet Orbiting Sun')
    fig.set_facecolor('black')
    ax.set_facecolor('black') 
    ax.legend()

    plt.show()
    plt.savefig("planeta.png")

'''
    pl_rade = radius_planet
    st_rad = radius_sun
    pl_orbsmax = radius_orb    
'''
```
